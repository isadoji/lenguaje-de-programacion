Lista de Ejercicios para Tarea en Casa del Módulo 1
Ejercicios
Ejercicio 1: Saludo Personalizado

Escribe un programa que pida al usuario su nombre y lo salude con un mensaje personalizado.
Entrada: Nombre del usuario (cadena de texto).
Salida: Mensaje de saludo personalizado.
Ejercicio 2: Suma de Dos Números

Escribe un programa que pida al usuario dos números y calcule su suma.
Entrada: Dos números (flotantes).
Salida: La suma de los dos números.
Ejercicio 3: Conversión de Temperatura

Escribe un programa que convierta una temperatura dada en grados Celsius a grados Fahrenheit.
Fórmula: ( F = \frac{9}{5}C + 32 )
Entrada: Temperatura en grados Celsius (flotante).
Salida: Temperatura en grados Fahrenheit.
Ejercicio 4: Cálculo del Área de un Círculo

Escribe un programa que calcule el área de un círculo dado su radio.
Fórmula: ( A = \pi r^2 )
Entrada: Radio del círculo (flotante).
Salida: Área del círculo.
Ejercicio 5: Promedio de Notas

Escribe un programa que pida al usuario cinco notas y calcule su promedio.
Entrada: Cinco notas (flotantes).
Salida: Promedio de las notas.
Ejercicio 6: Conversión de Unidades

Escribe un programa que convierta una distancia dada en kilómetros a millas.
Fórmula: ( 1 \text{ km} = 0.621371 \text{ millas} )
Entrada: Distancia en kilómetros (flotante).
Salida: Distancia en millas.
Ejercicio 7: Cálculo del Perímetro de un Rectángulo

Escribe un programa que calcule el perímetro de un rectángulo dadas su longitud y ancho.
Fórmula: ( P = 2(l + w) )
Entrada: Longitud y ancho del rectángulo (flotantes).
Salida: Perímetro del rectángulo.
Ejercicio 8: Cálculo del Volumen de una Esfera

Escribe un programa que calcule el volumen de una esfera dado su radio.
Fórmula: ( V = \frac{4}{3} \pi r^3 )
Entrada: Radio de la esfera (flotante).
Salida: Volumen de la esfera.
Ejercicio 9: Cálculo del Índice de Masa Corporal (IMC)

Escribe un programa que calcule el IMC de una persona dada su altura y peso.
Fórmula: ( IMC = \frac{peso}{altura^2} )
Entrada: Peso (kg) y altura (m) de la persona (flotantes).
Salida: IMC de la persona.
Ejercicio 10: Conversión de Tiempo

Escribe un programa que convierta una cantidad de tiempo dada en minutos a horas y minutos.
Entrada: Tiempo en minutos (entero).
Salida: Tiempo en horas y minutos.
