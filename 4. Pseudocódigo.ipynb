# 4. Pseudocódigo y diagramas de flujo

## ¿Qué es un Algoritmo?

Un **algoritmo** es una secuencia finita de pasos bien definidos que llevan a la solución de un problema.

Ejemplo: Un algoritmo para hacer un café podría incluir los siguientes pasos:

1. Calentar agua.
2. Colocar café en un filtro.
3. Verter el agua caliente sobre el café.
4. Servir el café en una taza.
5. Agregar azúcar o leche si se desea.


## ¿Qué es el Pseudocódigo?


El **pseudocódigo** es una forma de escribir algoritmos de manera estructurada sin usar un lenguaje de programación específico.

Sirve como puente entre la solución del problema y su implementación en un lenguaje como Python.

Ejemplo de pseudocódigo para saber si un número es par o impar:

    Inicio
        Escribir "Ingrese un número:"
        Leer num
        Si num % 2 == 0 Entonces
        Escribir "El número es par"
        Si no
        Escribir "El número es impar"
    FinSi
    Fin


## Sintaxis Básica del Pseudocódigo


- Uso de **Inicio** y **Fin** para delimitar el algoritmo.
- Uso de **Leer** para capturar datos del usuario.
- Uso de **Escribir** para mostrar resultados.
- Uso de variables para almacenar datos.
- Operadores matemáticos y estructuras de control como condicionales y bucles.

## Traducción de Pseudocódigo a Python


```python
num = int(input("Ingrese un número: "))
if num % 2 == 0:
    print("El número es par")
else:
    print("El número es impar")
```

## ¿Qué es un diagrama de flujo?

- Es una representación gráfica de un algoritmo o proceso.
- Se utilizan símbolos estandarizados para ilustrar la secuencia de pasos de un programa.

Los elementos básicos de un diagrama de flujo son:

1. **Óvalo:** Representa el inicio o fin del algoritmo.
2. **Paralelogramo:** Indica entrada o salida de datos.
3. **Rectángulo:** Representa un proceso o acción.
4. **Rombo:** Expresa una decisión (sí/no).
5. **Flechas:** Indican el flujo de ejecución.

Para generar diagramas de flujo en Python, utilizamos la librería `graphviz`.

    !pip install graphviz

```python
import graphviz
    
'''  
Creación del Diagrama de Flujo en Python
'''

def crear_diagrama():   
    dot = graphviz.Digraph()

    '''  
    Cada nodo representa un paso en el diagrama de flujo:
        A. Inicio del algoritmo (óvalo).
        B. Lectura de un número (paralelogramo).
        C. Verificación de si el número es par (rombo, decisión).
        D. Acción si el número es par (rectángulo).
        E. Acción si el número es impar (rectángulo).
        F. Fin del algoritmo (óvalo).
    '''
    dot.node('A', 'Inicio', shape='oval')
    dot.node('B', 'Leer número', shape='parallelogram')
    dot.node('C', 'Número par?', shape='diamond')
    dot.node('D', 'Imprimir "Es par"', shape='rectangle')
    dot.node('E', 'Imprimir "Es impar"', shape='rectangle')
    dot.node('F', 'Fin', shape='oval')
    
    '''  
    Se agregan flechas entre los nodos para definir el flujo del algoritmo.
    
    label='Sí' y label='No' indican las decisiones en la estructura condicional.  
    '''
    
    dot.edge('A', 'B')
    dot.edge('B', 'C')
    dot.edge('C', 'D', label='Sí')
    dot.edge('C', 'E', label='No')
    dot.edge('D', 'F')
    dot.edge('E', 'F')
    
    '''  
    La función muestra el diagrama de flujo generado.
    crear_diagrama() se ejecuta para mostrar el resultado.
    '''
    return dot

crear_diagrama()
```

## Traducción de Pseudocódigo a Python

### 2.1. Entrada y Salida

    Pseudocódigo:
        Leer número
        Escribir "El número ingresado es:", número


### En Python:
    n = int(input("Ingrese un número: "))
    print("El número ingresado es:", n)

### 2.2. Condicionales

    Pseudocódigo:
        Si número es par entonces
            Escribir "Es par"
        Sino
            Escribir "Es impar"


### En Python:
    if n % 2 == 0:
        print("Es par")
    else:
        print("Es impar")

### 2.3. Bucles

    Pseudocódigo (bucle para imprimir los primeros 5 números):
        Para i desde 1 hasta 5 hacer
            Escribir i
        FinPara


### En Python:
    for i in range(1, 6):
        print(i)
